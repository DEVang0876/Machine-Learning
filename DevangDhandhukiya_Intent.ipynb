```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import warnings
from sklearn.preprocessing import StandardScaler,LabelEncoder
from sklearn.model_selection import train_test_split,cross_val_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.decomposition import PCA
```

```python
le = LabelEncoder()

df['Month'] = le.fit_transform(df['Month'])
df['VisitorType'] = le.fit_transform(df['VisitorType'])
df['Weekend'] = le.fit_transform(df['Weekend'])
df['Revenue'] = le.fit_transform(df['Revenue'])
```

```python
df['Revenue'].value_counts()
```

```python
X = df.drop(columns='Revenue',axis=1)
y = df['Revenue']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
sc = StandardScaler()
X_train_sc = sc.fit_transform(X_train)
X_test_sc = sc.fit_transform(X_test)
```

```python
from sklearn.metrics import *
```

## Logistics Regression

```python
model_lr = LogisticRegression(max_iter=100)
model_lr.fit(X_train_sc,y_train)
y_pred_lr = model_lr.predict(X_test_sc)
model_lr_acc = accuracy_score(y_test,y_pred_lr)
model_lr_acc
cm=confusion_matrix(y_test, y_pred_lr)
cm
sns.heatmap(cm, annot=True, cmap='copper')
```

## KNN

```python
model_knn = KNeighborsClassifier(n_neighbors=5)
model_knn.fit(X_train_sc,y_train)
y_pred_knn = model_knn.predict(X_test_sc)
model_knn_acc =  accuracy_score(y_test,y_pred_knn)
model_knn_acc
cm=confusion_matrix(y_test, y_pred_knn)
cm
sns.heatmap(cm, annot=True, cmap='copper')
```

## RandomForest

```python
model_rf = RandomForestClassifier(n_estimators=100)
model_rf.fit(X_train_sc,y_train)
y_pred_rf=model_rf.predict(X_test_sc)
model_rf_acc=accuracy_score(y_test,y_pred_rf)
model_rf_acc
cm=confusion_matrix(y_test, y_pred_rf)
cm
sns.heatmap(cm, annot=True, cmap='copper')
```

```python
X_train_scaled = sc.fit_transform(X_train)
X_test_scaled = sc.fit_transform(X_test)
```

##  Train a Logistic Regression Model and Make Predictions

```python

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Train logistic regression model
log_model = LogisticRegression(max_iter=1000)
log_model.fit(X_train_scaled, y_train)

# Predict on test data
y_pred_log = log_model.predict(X_test_scaled)

# Evaluate model
log_acc = accuracy_score(y_test, y_pred_log)
print("Logistic Regression Accuracy:", log_acc)
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred_log))
print("Classification Report:\n", classification_report(y_test, y_pred_log))
```

##  Add Random Noise and Evaluate KNN

```python

import numpy as np
from sklearn.neighbors import KNeighborsClassifier

# Add 10 columns of random noise
np.random.seed(42)
noise_train = np.random.rand(X_train_scaled.shape[0], 10)
noise_test = np.random.rand(X_test_scaled.shape[0], 10)

X_train_noise = np.hstack((X_train_scaled, noise_train))
X_test_noise = np.hstack((X_test_scaled, noise_test))

# Train KNN on noisy data
knn_noise = KNeighborsClassifier(n_neighbors=5)
knn_noise.fit(X_train_noise, y_train)
y_pred_knn_noise = knn_noise.predict(X_test_noise)

# Evaluate
knn_noise_acc = accuracy_score(y_test, y_pred_knn_noise)
print("KNN Accuracy with Noise:", knn_noise_acc)
```

## Step 8: Evaluate KNN Accuracy for Different Values of k

```python

import matplotlib.pyplot as plt

train_acc = []
test_acc = []
k_vals = range(1, 51)

for k in k_vals:
    model = KNeighborsClassifier(n_neighbors=k)
    model.fit(X_train_scaled, y_train)
    train_acc.append(model.score(X_train_scaled, y_train))
    test_acc.append(model.score(X_test_scaled, y_test))

plt.figure(figsize=(10, 5))
plt.plot(k_vals, train_acc, label='Training Accuracy')
plt.plot(k_vals, test_acc, label='Testing Accuracy')
plt.xlabel('k (Number of Neighbors)')
plt.ylabel('Accuracy')
plt.title('KNN Accuracy for Different k')
plt.legend()
plt.grid(True)
plt.savefig("knn_k_vs_accuracy.png")
plt.show()
```

##  Cross-Validation with KNN (k=5)

```python

from sklearn.model_selection import cross_val_score

knn_cv = KNeighborsClassifier(n_neighbors=5)
cv_scores = cross_val_score(knn_cv, X_train_scaled, y_train, cv=10)
print("Cross-Validation Accuracies:", cv_scores)
print("Average CV Accuracy:", np.mean(cv_scores))
```

##  PCA for Dimensionality Reduction and KNN

```python

from sklearn.decomposition import PCA

# Apply PCA
pca = PCA(n_components=5)
X_train_pca = pca.fit_transform(X_train_scaled)
X_test_pca = pca.transform(X_test_scaled)

# Train KNN with PCA-reduced features
knn_pca = KNeighborsClassifier(n_neighbors=5)
knn_pca.fit(X_train_pca, y_train)
y_pred_pca = knn_pca.predict(X_test_pca)

# Evaluate
pca_knn_acc = accuracy_score(y_test, y_pred_pca)
print("KNN with PCA Accuracy:", pca_knn_acc)
```

##  Handle Class Imbalance with SMOTE and Train KNN

```python

from imblearn.over_sampling import SMOTE

# Apply SMOTE
smote = SMOTE(random_state=42)
X_train_bal, y_train_bal = smote.fit_resample(X_train_scaled, y_train)

# Train KNN
knn_smote = KNeighborsClassifier(n_neighbors=5)
knn_smote.fit(X_train_bal, y_train_bal)
y_pred_smote = knn_smote.predict(X_test_scaled)

# Evaluate
knn_smote_acc = accuracy_score(y_test, y_pred_smote)
print("KNN with SMOTE Accuracy:", knn_smote_acc)
```
